univariate categorical analysis

```python
cat_columns=[]
for i in df.columns:
  if (df[i].nunique()) <10:
    cat_columns.append(i)
```

```python
cat_columns
```

```python
fig,axes=plt.subplots(nrows=16,ncols=2,figsize=(20,60))
for i in range(0,len(cat_columns)):
  for j in range(0,2):

    if j==1:

      df[cat_columns[i]].value_counts().plot(kind='bar',ax=axes[i,j],title=f"{cat_columns[i]}--> barchart")
    else:
       df[cat_columns[i]].value_counts().plot(kind='pie',autopct='%.2f%%',ax=axes[i,j],title=f"{cat_columns[i]}-->piechart")



  i+=1

plt.tight_layout()
plt.show()
```

multivariate analysis


```python
#categoriacl and numerical analysis
```

```python
fig,axes=plt.subplots(nrows=6,ncols=1,figsize=(8,40))
for i in range(len(numerical_col)):
  sns.stripplot(data=df,x='Class',y=numerical_col[i],ax=axes[i],jitter=True)
  axes[i].set_title(f'strippplot Class category vs {numerical_col[i]}')

  i+=1
plt.tight_layout()
plt.show()
```

```python
s=df.groupby([df['v2'],df['Class']])
```

```python
v=s.size().unstack()
```

```python
v.plot(kind='bar',stacked=True)
```

```python
#category vs category
```

```python
fig,axes=plt.subplots(nrows=4,ncols=4,figsize=(8,20))
for i in range(4):
  for j in range(4):
    index=i+j
    df.groupby([f'{cat_columns[index]}','Class']).size().unstack().plot(kind='bar',stacked=True,title=f"{cat_columns[index]} vs Class",ax=axes[i,j])
plt.subplots_adjust(left=0.2, right=1.5, top=0.9, bottom=0.1, wspace=1, hspace=0.4)

```

```python
#relation ship between numerical columns by heatmap  not including target column as it is a categorical in nature

corr=df[numerical_col].corr()
sns.heatmap(data=corr,annot=True)
```
